# TerraScore: Google Earth Engine Dataset to CSV

This notebook validates Earth Engine access and exports a sample table to CSV after verification.

```python
import ee

try:
    ee.Initialize()
except Exception:
    ee.Authenticate()
    ee.Initialize()

print('Earth Engine initialized successfully')
```

```python
dataset = ee.ImageCollection('MODIS/061/MOD13Q1').first()
region = ee.Geometry.Rectangle([92.0, 24.0, 96.0, 28.0])

sample = dataset.sample(region=region, scale=250, numPixels=50, geometries=False)

task = ee.batch.Export.table.toDrive(
    collection=sample,
    description='terrascore_gee_sample_export',
    fileFormat='CSV'
)
task.start()
print('Export started. Check task status in Earth Engine Tasks panel.')
```
